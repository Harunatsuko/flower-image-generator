```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from flower import Flower
```

```python
# for i in np.arange(0,10,1):
flower = Flower(40,300)
level_cnt = np.random.randint(2,7)
petal_kinds = [np.random.randint(1,7) for i in range(level_cnt)]
if petal_kinds[0] in {1,4}:
    grad_colors = 'pink'
elif petal_kinds[0] in {2,5}:
    grad_colors = 'hot'
else:
    grad_colors = 'cold'
flower.draw(level_cnt,petal_kinds=petal_kinds,grad_colors=grad_colors)
img = flower.img*255
img = np.float32(img)
plt.figure(figsize=(8, 8))
plt.imshow(flower.img)
#     img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
#     cv2.imwrite(str(i)+'.png',img)
```
